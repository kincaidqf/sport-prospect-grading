# NBA Draft Prospect — Exploratory Data Analysis

Goals:
1. Understand NCAA stat distributions and missingness
2. Explore NBA target variable (PLUS_MINUS) distribution
3. Check scouting report coverage and text length distribution
4. Identify feature columns to populate `STAT_FEATURE_COLS` in `preprocessor.py`

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
# Load data
ncaa = pd.read_csv('../data/ncaa/ncaa_master.csv')
nba  = pd.read_csv('../data/nba/nba_master.csv')
players = pd.read_csv('../data/scouting/players.csv')

print('NCAA:', ncaa.shape)
print('NBA: ', nba.shape)
print('Players (scouting):', players.shape)
```

```python
# Column inventory
print('=== NCAA columns ==='); print(ncaa.columns.tolist())
print('\n=== NBA columns ==='); print(nba.columns.tolist())
print('\n=== Players columns ==='); print(players.columns.tolist())
```

```python
# Missingness
print('NCAA missing:\n', ncaa.isnull().mean().sort_values(ascending=False).head(20))
```

```python
# NBA target distribution
fig, ax = plt.subplots(figsize=(8, 4))
nba['PLUS_MINUS'].hist(bins=50, ax=ax)
ax.set_title('NBA PLUS_MINUS distribution')
ax.set_xlabel('PLUS_MINUS')
plt.tight_layout()
```

```python
# Scouting report text lengths
with open('../data/scouting/scouting_texts.txt') as f:
    texts = f.read().strip().split('\n\n')

lengths = [len(t.split()) for t in texts]
print(f'Reports: {len(texts)}')
print(f'Words — mean: {np.mean(lengths):.0f}, median: {np.median(lengths):.0f}, max: {max(lengths)}')

pd.Series(lengths).hist(bins=40)
plt.title('Scouting report word count distribution')
plt.xlabel('words')
plt.tight_layout()
```
